```python
import requests
from bs4 import BeautifulSoup

def scrape_web_content(urls):
  web_data = {}
  for url in urls:
    try:
      response = requests.get(url)
      soup = BeautifulSoup(response.content, "html.parser")
      # Εξαγωγή τίτλου και παραγράφων
      title = soup.title.string if soup.title else "No Title"
      paragraphs = soup.find_all('p')
      content = " ".join([p.get_text() for p in paragraphs])

      web_data[url] = {
          "title": title,
          "content": content
      }
      print(f"Δεδομένα εξήχθησαν από το URL: {url}")
    except Exception as e:
      print(f"Σφάλμα κατά την εξαγωγή δεδομένων από το URL {url}: {e}")
  return web_data


web_urls = [
"https://www.greekmythology.com/",
"https://www.hellenicmythology.gr/",
]


web_content = scrape_web_content(web_urls)

output_path = "/content/drive/MyDrive/ISA Workshop/web_transcripts.json"

# Save the transcripts to Google Drive
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(youtube_transcripts, f, ensure_ascii=False, indent=4)

print(f"Transcripts saved to: {output_path}")
```

```python
import re

def preprocess_text(text):
  """
  Preprocess the input text by:
  - Removing line breaks within words
  - Removing standalone line breaks
  - Replacing multiple spaces with a single space
  - Normalising multiple dots into a single dot
  """
  text = text.replace("-\n", "")
  text = text.replace("\n", " ")
  text = re.sub(r"\s+", " ", text)
  text = re.sub(r"\.+", ".", text)

  return text
```

```python
def preprocess_dataset(input_file, output_file):
  """
  Προεπεξεργάζεται τα δεδομένα από το αρχείο JSON και τα αποθηκεύει σε νέο αρχείο.
  """
  with open(input_file, "r", encoding="utf-8") as f:
    dataset = json.load(f)


  for entry in dataset:
    entry["text"] = preprocess_text(entry["text"])


  with open(output_file, "w", encoding="utf-8") as f:
    json.dump(dataset, f, ensure_ascii=False, indent=4)

data_path = "/content/drive/MyDrive/ISA Workshop/final_dataset.json"
output_path = "/content/drive/MyDrive/ISA Workshop/preprocessed_dataset.json"
preprocess_dataset(data_path, output_path)
```

```python
import spacy
from tqdm import tqdm

# Load English and Greek SpaCy models
nlp_el = spacy.load("el_core_news_sm")

def chunk_text(text, spacy_nlp, chunk_size=128):
  """
  Hybrid chunking:
  1. Split text into sentences using SpaCy.
  2. Merge consecutive sentences until reaching the chunk size.
  """
  print("Splitting document into sentences using SpaCy...")
  sentences = [sent.text.strip() for sent in spacy_nlp(text).sents]
  print(f"Document contains {len(sentences)} sentences.")

  print(f"Merging sentences into chunks of max size {chunk_size} tokens...")
  chunks = [sentences[0]]
  for index in tqdm(range(1, len(sentences)), desc="Chunking text"):
    potential_chunk = " ".join([chunks[-1], sentences[index]])
    num_tokens_in_chunk = len([token for token in spacy_nlp(potential_chunk)])
    if num_tokens_in_chunk < chunk_size:
      chunks[-1] = potential_chunk
    else:
      chunks.append(sentences[index])

  print(f"Created {len(chunks)} chunks.")
  return chunks

def process_dataset(input_file, output_file, spacy_nlp, chunk_size=128):
    """
    Process a JSON dataset with text fields and apply chunking.
    """
    # Load dataset
    with open(input_file, "r", encoding="utf-8") as f:
      dataset = json.load(f)

    # Process each entry
    for entry in tqdm(dataset, desc="Processing dataset"):
      entry["chunks"] = chunk_text(entry["text"], spacy_nlp, chunk_size)

    # Save processed dataset
    with open(output_file, "w", encoding="utf-8") as f:
      json.dump(dataset, f, ensure_ascii=False, indent=4)

    print(f"Processed dataset saved to {output_file}")
```

```python
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")

def embed_chunks(chunks):
  """
  Δημιουργία embeddings για λίστα κειμένων.
  """
  return embedding_model.encode(chunks, show_progress_bar=True)

def embed_dataset(input_file, output_file):
  with open(input_file, "r", encoding="utf-8") as f:
    dataset = json.load(f)

  for entry in tqdm(dataset, desc="Embedding dataset"):
    entry["embeddings"] = embed_chunks(entry["chunks"]).tolist()

  with open(output_file, "w", encoding="utf-8") as f:
    json.dump(dataset, f, ensure_ascii=False, indent=4)

chunk_file = "/content/drive/MyDrive/ISA Workshop/chunked_dataset_el.json"
embedded_file = "/content/drive/MyDrive/ISA Workshop/embedded_dataset_el.json"
embed_dataset( chunk_file, embedded_file)
```

```python
import chromadb
from chromadb.api.types import Documents, EmbeddingFunction, Embeddings
import torch

class MultilingualSentenceTransformer(EmbeddingFunction):
    """Create a ChromaDB embedding function based on the chosen model."""
    def __init__(self, model="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"):
        self.model = model
        self.embedder = self.initialize_model(model)

    def initialize_model(self, model):
        from sentence_transformers import SentenceTransformer
        embedder = SentenceTransformer(model, device="cuda" if torch.cuda.is_available() else "cpu")
        return embedder

    def __call__(self, sentences: Documents) -> Embeddings:
        return self.embedder.encode(sentences, convert_to_numpy=True).tolist()

def batchify(iterable, n=1):
    """Split iterable into batches of size n."""
    l = len(iterable)
    for ndx in range(0, l, n):
        yield iterable[ndx: min(ndx + n, l)]

def create_and_save_collection(dataset_file, collection_name="rag_collection", batch_size=12):
    """
    Create and save a ChromaDB collection to Google Drive.
    """
    print(f"Loading dataset from {dataset_file}...")
    with open(dataset_file, "r", encoding="utf-8") as f:
        dataset = json.load(f)

    print("Extracting chunks for embedding")
    chunks = []
    metadata = []
    for entry in dataset:
        chunks.extend(entry["chunks"])
        metadata.extend([{"source": entry["source"]}] * len(entry["chunks"]))

    print("Creating and indexing vector store")
    client = chromadb.PersistentClient(path=COLLECTION_SAVE_PATH)  # Persistent storage
    embedding_function = MultilingualSentenceTransformer()
    collection = client.create_collection(
        collection_name,
        metadata={"hnsw:space": "cosine"},
        embedding_function=embedding_function,
    )
    id_num = 0
    for batch in tqdm(batchify(chunks, n=batch_size), desc="Indexing vector database", total=len(chunks) // batch_size):
        texts = batch
        collection.add(
            documents=texts,
            ids=[f"id{idx}" for idx in range(id_num, id_num + len(texts))],
            metadatas=metadata[id_num: id_num + len(texts)],
        )
        id_num += len(texts)

    print(f"Collection '{collection_name}' created and saved at {COLLECTION_SAVE_PATH}.")
    return collection

def load_collection():
    """
    Load a ChromaDB collection from Google Drive.
    """
    client = chromadb.PersistentClient(path=COLLECTION_SAVE_PATH)
    return client
```

```python
import chromadb
from chromadb.api.types import Documents, EmbeddingFunction, Embeddings
import torch
from openai import OpenAI
import json
from tqdm import tqdm

# Define paths for saving the collection and embeddings
COLLECTION_SAVE_PATH = "/content/drive/MyDrive/ISA Workshop/collection_el"

# API setup for Meltemi
MELTEMI_API_KEY = "sk-RYF0g_hDDIa2TLiHFboZ1Q"
MELTEMI_BASE_URL = "http://ec2-3-19-37-251.us-east-2.compute.amazonaws.com:4000/"
MELTEMI_CLIENT = OpenAI(api_key=MELTEMI_API_KEY, base_url=MELTEMI_BASE_URL)

# Prompt Template
PROMPT_TEMPLATE = """Passages:\n\n{passages}\n\nQuery: {query}"""
RAG_SYSTEM_PROMPT = """Instructions: Compose a comprehensive reply to the query, based on the provided passages.
If the search results mention multiple subjects with the same name, create separate answers for each.
In your answers, maintain the appropriate knowledge level for a person that is consuming this material.
You MAY include appropriate extra information for providing clear, simple and comprehensive clarifications,
if the user explicitly asks for it, else you only respond based on the provided passages.
Do not make the responses neither too advanced, nor too simplistic.
If the text does not relate at all to the query, propose more appropriate prompt queries based on the book contents.
Provide your answer in the same language as the user's query."""

def retrieve_relevant_passages(query, chroma_collection, n_results=4, max_length=1000):
  """Retrieve relevant passages and truncate their length."""
  print(f"Querying for: {query}")
  result = chroma_collection.query(
    query_texts=[query],
    n_results=n_results,
  )
  documents = result.get("documents", [])
  metadatas = result.get("metadatas", [])
  # Truncate passages
  truncated_docs = [doc[:max_length] for doc in documents[0]] if documents else []
  return list(zip(truncated_docs, metadatas[0])) if truncated_docs else []

def format_prompt_with_sources(query, chroma_collection, n_neighbors=5):
  """Format the prompt with passages and metadata retrieved from the ChromaDB collection."""
  results = chroma_collection.query(query_texts=[query], n_results=n_neighbors)

  if not results.get("documents"):
    print("No relevant documents found. Returning original query.")
    return query, []

  passages = "\n\n".join(results["documents"][0])
  sources = {metadata.get("source", "Unknown") for metadata in results["metadatas"][0]}
  prompt = PROMPT_TEMPLATE.format(passages=passages, query=query)
  return prompt, sources

def query_meltemi(prompt):
  """Query the Meltemi model with the formatted prompt."""
  response = MELTEMI_CLIENT.chat.completions.create(
    model="meltemi",
    messages=[
      {"role": "system", "content": RAG_SYSTEM_PROMPT},
      {"role": "user", "content": prompt},
    ],
    max_tokens=3000,
  )
  return response.choices[0].message.content

def rag_query_meltemi_with_sources(query, collection_name, client, n_results=5, max_length=1000):
  """Retrieve passages using RAG and generate a response with the Meltemi model, including sources."""
  collection = client.get_collection(collection_name)
  # Retrieve passages with truncation
  passages = retrieve_relevant_passages(query, collection, n_results=n_results, max_length=max_length)
  if not passages:
    return "No relevant passages found.", []

  # Format the prompt
  formatted_passages = "\n\n".join([p[0] for p in passages])
  sources = {p[1].get("source", "Unknown") for p in passages}
  prompt = PROMPT_TEMPLATE.format(passages=formatted_passages, query=query)

  # Query Meltemi
  response = query_meltemi(prompt)

  # Append sources
  sources_text = "\n\nΠηγές:\n" + "\n".join(sources) if sources else "\n\nΠηγές: Καμία."
  final_response = response + sources_text

  print("Query:")
  print(query)
  print("\nRAG-Enhanced Prompt:")
  print(prompt)
  print("\nResponse:")
  print(final_response)
  print("--------------------------------------")
  return final_response

def query_greek_with_rag(query, collection_name=COLLECTION_SAVE_PATH, n_results=5, max_length=500):
  """Query the Greek dataset using RAG and the Meltemi model."""
  client = chromadb.PersistentClient(path=COLLECTION_SAVE_PATH)
  response = rag_query_meltemi_with_sources(query, collection_name, client, n_results=n_results, max_length=max_length)
  return response

# Save the collection in Google Drive for persistent use
COLLECTION_NAME_EL = "collection_el"
query_greek_with_rag("Πες μου ένα μύθο", collection_name=COLLECTION_NAME_EL, n_results=4)
```
